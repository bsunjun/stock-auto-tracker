# ThemeRadar Starter (skeleton)

이 notebook 은 **skeleton 입니다**. 실행 가능한 ETL 코드를 포함하지 않습니다.

- Telegram / 뉴스 raw 텍스트 정규화 보조용.
- 산출은 `theme_radar_pack` 초안.
- 사람 검토 후 사람이 Drive 저장 (자동 저장 금지).

## 절대 금지

- OCR / Vision / 외부 LLM API client 미사용.
- 사용자 ID / 비밀번호 / 세션 토큰 / 인증 쿠키 저장 금지.
- 메시지 raw 본문을 출력 셀에 그대로 남기지 않을 것.
- Drive 자동 write 미수행.


## 1. 입력 정의 (수동 입력)

사람이 INBOX 의 요약 텍스트를 아래 변수에 직접 붙여넣습니다.
메시지 raw 전문은 붙여넣지 않고 **요약 + 시각 + 채널 ID** 만 입력합니다.

## 2. 정규화 outline

- KST 시각 정규화 (timezone-aware).
- 채널 ID 해시화 (개인 식별정보 제거).
- 테마 후보 키워드 추출 (간단한 keyword bag, 외부 LLM 호출 없음).
- 채널 sync count 계산.
- velocity score 산출 (간단한 카운트 변화율).

## 3. 산출 outline

- `theme_radar_pack` 스키마(`schemas/theme_radar_pack.schema.json`)의 필드 일부 채움:
  - `themes[].theme_id`
  - `themes[].theme_label`
  - `themes[].channel_sync_count`
  - `themes[].velocity_score`
  - `themes[].paid_only`
  - `themes[].pb_state_recommendation` (`WATCH_ONLY` 또는 `THEME_DISCOVERY` 만)
- 사람이 검토 후 최종 확정.

## 4. Hard rules (Colab 측)

- `direct_trade_signal=false`, `trade_signal in (null, false)`.
- Telegram-only 단서로 PB_READY 이상 권장 금지.
- 메시지 raw 텍스트 출력 금지.
- API key / token / password 저장 금지.
