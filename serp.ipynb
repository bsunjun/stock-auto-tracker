```python
from serpapi import GoogleSearch

params = {
    "engine": "google_finance",
    "q": "KRX:005930",
    "api_key": "ca025ef8e8b59f79008f7c57271c420b732842111cd241fa75d8c239b2560945"
}

search = GoogleSearch(params)
result = search.get_dict()

# price 추출
try:
    price_info = result["futures_chain"][0]  # 첫 번째 결과
    name = price_info["date"]  # 실제로는 종목명
    price = price_info["price"]
    change = price_info["price_movement"]["percentage"]
    print(f"{name} 현재가: {price}원 ({change:+.2f}%)")
except Exception as e:
    print("데이터 추출 실패:", e)
```
